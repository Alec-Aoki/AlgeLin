#  Matrizes úteis para referência

* **Reflexão sobre o eixo $x$:**
  
 $ R_x = $
  $\begin{bmatrix}
  1 & 0 \\
  0 & -1
  \end{bmatrix}
  $

* **Reflexão na reta $y = x$:**
 $
  R_{y=x} = $
  $\begin{bmatrix}
  0 & 1 \\
  1 & 0
  \end{bmatrix}
  $

* **Rotação de $\theta$ graus anti-horária:**
 $
  R_\theta =$
 $ \begin{bmatrix}
  \cos \theta & -\sin \theta \\
  \sin \theta & \cos \theta
  \end{bmatrix}
  $

* **Translação por vetor $\vec{v} = (a, b)$:**
  $
  T(\vec{x}) = \vec{x} + \vec{v}
  $

```python
# Import das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

# Exercício 1
Considere o triângulo  $ABC$ , com vértices definidos por:

*  A = (1, 1)
*  B = (3, 1)
*  C = (2, 3)

Represente o triângulo como uma matriz, onde cada **coluna** é um ponto:


T=
$\begin{bmatrix}
1 & 3 & 2 \\
1 & 1 & 3
\end{bmatrix}$



 Escreva as **matrizes** das seguintes transformações e aplique **individualmente** cada transformação ao triângulo $T$, utilizando:

 Multiplicação de matrizes $T' = A \cdot T$, para isometrias lineares e soma vetorial $T' = A \cdot T + \vec{v}$, quando houver translação

  Visualize (graficamente) o triângulo original e cada triângulo transformado.

```python
# Triângulo original
pontos_triangulo = np.array([[1,1], [3,1], [2,3]])

fig, ax = plt.subplots()

triangulo = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)

ax.add_patch(triangulo)

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

   a) **Reflexão** em relação ao eixo $x$

```python
fig, ax = plt.subplots()

# Triângulo original para referência
triangulo2 = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo2)

# Refletir em relação ao eixo é simplesmente inverter o sinal da coordenada y de cada ponto
reflexao_eixoX = np.array([[1,0],[0,-1]])
pontos_reflexaoEixoX = pontos_triangulo @ reflexao_eixoX
print('Matriz de reflexao: \n', pontos_reflexaoEixoX)

triangulo_reflexaoEixoX = patches.Polygon(pontos_reflexaoEixoX, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_reflexaoEixoX)

ax.set_xlim(0, 4)
ax.set_ylim(-4, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

   b) **Reflexão** em relação à reta $y = x$

```python
fig, ax = plt.subplots()

# Triângulo original para referência
triangulo = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo)

# Reta y = x para referência
x = np.linspace(0, 4, 100)
y = x
ax.plot(x, y, color='black', linewidth=1)

# Inverter as coordenadas y e x de cada ponto
reflexao_reta = np.array([[0,1],[1,0]])
pontos_reflexaoReta = pontos_triangulo @ reflexao_reta
print('Matriz reflexao na reta: \n', pontos_reflexaoReta)

triangulo_reflexaoReta = patches.Polygon(pontos_reflexaoReta, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_reflexaoReta)

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

   c) **Rotação** no sentido anti-horário em torno da origem por um ângulo genérico $\theta$ (em graus)

Dica: use a fórmula

  $R_\theta = $
  $$
  \begin{bmatrix}
  \cos(\theta) & -\sin(\theta) \\
  \sin(\theta) & \cos(\theta)
  \end{bmatrix}
  $$

```python
def matriz_rotacao (t):
    return [[np.cos(t), -(np.sin(t))], [np.sin(t), np.cos(t)]]
```

```python
fig, ax = plt.subplots()

# Triângulo original para referência
triangulo = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo)

# Rotação de 180 graus
theta = np.radians(180)
print('Matriz rotacao: \n', matriz_rotacao(theta))
pontos_rotacao = pontos_triangulo @  matriz_rotacao(theta)

triangulo_rotacao = patches.Polygon(pontos_rotacao, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_rotacao)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

   d) **Translação** pelo vetor $\vec{v} = (2, -1)$

```python
fig, ax = plt.subplots()

# Triângulo original para referência
triangulo = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo)

# Somando o vetor a cada ponto
vetor = np.array([2, -1])
pontos_translacao = pontos_triangulo + vetor

triangulo_translacao = patches.Polygon(pontos_translacao, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_translacao)

ax.set_xlim(0, 6)
ax.set_ylim(-1, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Composição de Isometrias

 Considere a seguinte composição de transformações:

   * Primeiro: **rotação** de $\theta = 90^\circ$
   * Depois: **reflexão** no eixo $x$

   a) Escreva a **matriz composta** $M = R_x \cdot R_{90}$


```python
theta = np.radians(90)
matriz_composta = reflexao_eixoX @ matriz_rotacao(theta)

print('M:\n', matriz_composta)
```

b) Aplique $M$ ao triângulo $T$ e visualize o resultado.

```python
fig, ax = plt.subplots()

# Triângulo original para referência
triangulo = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo)

pontos_composta = pontos_triangulo @ matriz_composta
print('Pontos:\n', pontos_composta)

triangulo_composta = patches.Polygon(pontos_composta, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_composta)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

   c) Interprete geometricamente o efeito da transformação composta. O que ela faz com o triângulo?


Ela executou primeiro a rotação e em seguida a reflexão, combinando ambas as transformações em uma só.

 Escolha um valor arbitrário de $\theta$ (por exemplo, $45^\circ$) e repita os passos acima.

```python
theta = np.radians(45)
matriz_composta = reflexao_eixoX @ matriz_rotacao(theta)

print('M:\n', matriz_composta)

fig, ax = plt.subplots()

# Triângulo original para referência
triangulo6 = patches.Polygon(pontos_triangulo, closed=True, facecolor='lightblue', edgecolor='blue', linewidth=2)
ax.add_patch(triangulo6)

pontos_composta = pontos_triangulo @ matriz_composta
print('Pontos:\n', pontos_composta)

triangulo_composta = patches.Polygon(pontos_composta, closed=True, facecolor='red', edgecolor='darkred', linewidth=2)
ax.add_patch(triangulo_composta)

ax.set_xlim(-1, 4)
ax.set_ylim(-4, 4)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Exercício 3
## Análise Geométrica

 Após cada isometria ou composição:

   Justifique com código ou cálculo. Lembre-se: **isometrias preservam distâncias e ângulos**.

   a) As **distâncias** entre os vértices são preservadas?

```python
def distancias(pontos):
    d1 = np.linalg.norm(pontos[1] - pontos[0])  # distância entre vértice 0 e 1
    d2 = np.linalg.norm(pontos[2] - pontos[1])  # distância entre vértice 1 e 2
    d3 = np.linalg.norm(pontos[0] - pontos[2])  # distância entre vértice 2 e 0
    return d1, d2, d3
```

```python
print(np.allclose(distancias(pontos_triangulo), distancias(pontos_reflexaoEixoX)))
print(np.allclose(distancias(pontos_triangulo), distancias(pontos_reflexaoReta)))
print(np.allclose(distancias(pontos_triangulo), distancias(pontos_rotacao)))
print(np.allclose(distancias(pontos_triangulo), distancias(pontos_translacao)))
print(np.allclose(distancias(pontos_triangulo), distancias(pontos_composta)))
```

   b) Os **ângulos** internos permanecem os mesmos?

```python
def angulos(pontos):
    v1 = pontos[1] - pontos[0]
    v2 = pontos[2] - pontos[0]
    angulo = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
    return np.degrees(angulo)
```

```python
print(np.isclose(angulos(pontos_triangulo), angulos(pontos_reflexaoEixoX)))
print(np.isclose(angulos(pontos_triangulo), angulos(pontos_reflexaoReta)))
print(np.isclose(angulos(pontos_triangulo), angulos(pontos_rotacao)))
print(np.isclose(angulos(pontos_triangulo), angulos(pontos_translacao)))
print(np.isclose(angulos(pontos_triangulo), angulos(pontos_composta)))
```

OBS: temos que usar isclose() e allclose() pois são floats
