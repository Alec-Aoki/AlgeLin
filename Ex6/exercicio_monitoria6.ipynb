# Exercício 1

Considere, no espaço $\mathbb{R}^2$, as normas euclidiana $||.||_2$, da soma $||.||_1$ e do máximo $||.||_\infty$. Determine, geometricamente, para cada uma dessas normas, o conjunto $B=\{x\in \mathbb{R}^2 / ||x||=1\}$.




### **Exemplos de normas do $\mathbb{R}^n$**

$$\|\mathbf{x}\|_2 = \bigg( \sum_{i=1}^n |x_i|^2 \bigg)^{1/2}, \qquad
\|\mathbf{x}\|_\infty = \max \limits _{1 \leq i \leq n} | x_{i} |, \qquad
\|\mathbf{x}\|_1 = \sum _{i=1} ^n | x_{i} |$$

$\|\mathbf{x}\|_p = \bigg( \sum_{i=1}^n |x_i|^p \bigg)^{1/p}, \; p \geq 1$

Esfera com centro na origem e raio unitário.
$$S = \{\mathbf{x} \in \mathbb{R}^n; \; \|\mathbf{x}\|=1\}$$


```python
# Criar figura
plt.figure(figsize=(12, 4))
```

### Norma 2
$$ \|\mathbf{x}\|_2 = \bigg( \sum_{i=1}^n |x_i|^2 \bigg)^{1/2} = \sqrt{\sum_{i=1}^n |x_i|^2} = 1 $$

Como estamos em $\mathbb{R}^2$:

$ \sqrt{\sum_{i=1}^n |x_i|^2} = \sqrt{|x_1|^2 + |x_2|^2} = 1 $

$ |x_1|^2 + |x_2|^2 = 1 $

Essa é a equação para uma circuferência de raio 1 e centro em (0,0).

```python
# Parametrizando
theta = np.linspace(0, 2*np.pi, 100)
x1_norma2 = np.cos(theta)
x2_norma2 = np.sin(theta)

plt.plot(x1_norma2, x2_norma2, linewidth=2)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlabel('x1')
plt.ylabel('x2')
```

### Norma 1
$$ \|\mathbf{x}\|_1 = \sum _{i=1} ^n | x_{i} | = 1$$

Como estamos em $\mathbb{R}^2$:

$ |x_1| + |x_2| = 1 $

$ |x_2| = 1 - |x_1| $

Os pontos que resolvem esse sistema são: (0,1), (0,-1), (-1,0) e (1,0), resultando em um losango.


```python
# Pontos de x pra fazer as retas
x1_soma = np.linspace(-1, 1, 100)

x2_superior = 1 - np.abs(x1_soma) # Equação acima para x2 > 0
x2_inferior = -(1 - np.abs(x1_soma)) # Equação acima para x2 < 0

# Concatenando os pontos
x1_soma = np.concatenate([x1_soma, x1_soma[::-1]])
x2_soma = np.concatenate([x2_superior, x2_inferior[::-1]])

plt.plot(x1_soma, x2_soma, 'r-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlabel('x1')
plt.ylabel('x2')
```

### Norma Máxima
$$ \|\mathbf{x}\|_\infty = \max \limits _{1 \leq i \leq n} | x_{i} | = 1 $$

Como estamos em $\mathbb{R}^2$:

$ \max{(|x_1|, |x_2|)} = 1 $

Isso nos dá $|x_1| = \pm 1$ e $|x_2| = \pm 1$, gerando os pontos (-1,-1), (1,1), (-1,1) e (1,-1), formando um quadrado.

```python
x1_max = np.array([1, -1, -1, 1, 1])
x2_max = np.array([1, 1, -1, -1, 1])

plt.plot(x1_max, x2_max, 'g-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.axis('equal')
```

# Exercício 2
Considere uma base $B =  \{\mathbf{v}_1,\ldots,\mathbf{v}_{10}\}$ do $\mathbb{R}^{10}$ gerada pelo código abaixo. Implemente um código do processo de ortogonalização de Gram-Schmidt e obtenha uma base ortonormal do $\mathbb{R}^{10}$ a partir de $B$.

```python
# Garantir reprodutibilidade
np.random.seed(42)

# Começar com a matriz identidade
I = np.identity(10, dtype=int)

# Adicionar pequenas perturbações inteiras para evitar vetores ortogonais triviais
perturbation = np.random.randint(-3, 4, size=(10, 10))  # valores entre -3 e 3
B_int = I + perturbation

# Verificar se os vetores são linearmente independentes
rank = np.linalg.matrix_rank(B_int)
if rank == 10:
    # Imprimir vetores
    for i in range(10):
        print(f"v{i+1} = {B_int[:, i]}")
else:
    print("Os vetores não são linearmente independentes. Tente novamente.")


```

```python
# Matriz para armazenar a base ortonormal
U = np.zeros((10, 10)) # Inicializada com 0
    
for i in range(10):
    u_i = B_int[:, i].astype(float) # Pegando cada vetor de B
        
    # Subtraindo as projeções nos vetores
    for j in range(i):
        # Projeção de v_i em u_j
        proj = np.dot(u_i, U[:, j]) * U[:, j] # Produto interno * vetor
        u_i = u_i - proj
        
    # Normalizando o vetor e jogando na matriz
    U[:, i] = u_i / np.linalg.norm(u_i)
        
    print(f"\n u{i+1}:", U[:, i])
```

```python
print(f"\n u{i+1}:", U[:, i])
```

```python
# Verificar ortonormalidade
# U^T * U deve ser identidade
produto = U.T @ U
print(np.round(produto, 6))

print('\n')
    
# Verificar se os vetores são unitários
normas = [np.linalg.norm(U[:, i]) for i in range(10)]
print(np.round(normas, 6))
```
