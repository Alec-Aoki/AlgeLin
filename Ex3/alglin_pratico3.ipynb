Implemente uma visualização adequada para $U$, $V$ e  $U \cap V$.

```python
# Base da figura
fig = plt.figure() # Objeto figure
ax = plt.axes(projection='3d') # Objeto axes

# Nome dos eixos
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Meshgrid: grade retangular (coordenadas)
# Linspace: ticks de -10 a 10 de 5 em 5
s, t = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))

# Subespaço U
U = np.array([[0, 1, 0], [0, 0, 1]])

# Do Google AI:
# Para obter o plano a partir de dois vetores e um ponto de referência,
# use as equações paramétricas: P = tu + sv, 
# u e v são os vetores que geram o plano, e t e s são parâmetros escalares.

x_u = s * U[0][0] + t *  U[1][0]
y_u = s * U[0][1] + t *  U[1][1]
z_u = s * U[0][2] + t *  U[1][2]

ax.plot_surface(x_u, y_u, z_u, alpha=0.5, color='lightblue', label='U')
# alpha = transparencia

# Subespaço V
V = np.array([[1, 0, 0], [0, 0, 1]])

x_v = s * V[0][0] + t *  V[1][0]
y_v = s * V[0][1] + t * V[1][1]
z_v = s * V[0][2] + t *  V[1][2]

ax.plot_surface(x_v, y_v, z_v, alpha=0.5, color='lightgreen', label='V')

#######################################################

# Intersecção entre U e V
todos_vetores = np.vstack([U, V]) # Juntando os dois vetores em um só

# Retorna um vetor composto pelos vetores únicos do array, e a quantidade de
# cada vetor
vetores_unicos, quant = np.unique(todos_vetores, axis = 0, return_counts = True)
# Todo vetor que aparece mais de uma vez é a interseção
vetores_comuns = vetores_unicos[quant > 1]

for vetor in vetores_comuns:
    t = np.linspace(-10, 10, 20)

    x_intersec = t * vetor[0]
    y_intersec = t * vetor[1]
    z_intersec = t * vetor[2]

    ax.plot(x_intersec, y_intersec, z_intersec, color='red', label='Interseccao')
```

#### Exercício 02 - Transformação Linear
Seja a transformação linear $T:\mathbb{R}^2 \to \mathbb{R}^2$ associada a matriz
$$ A =
\begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
.$$
Considere a circunferência $C=\{(x,y)\in\mathbb{R}^2;\; x^2+y^2=1\}$. Implemente uma visualização da imagem da circunferência $C$ pela transformação linear $T$.
- Dica: Discretize um conjunto de pontos de $C$ através da parametrização $(x, y) = (cos(t), sen(t))$ e aplique a transformação nesses pontos.

### Solução
$T$ é dado por:
$ T(x,y) = A \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 2 && 1 \\ 1 && 2 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 2x + y \\ x + 2y \end{pmatrix} $

Parametrizando $C$:
$ \\
C: x^2 + y^2 = 1 \\
(x, y) = (\cos(t), \sin(t)), t \in [0, 2 \pi]
$

Aplicando $C$ em $T$:
$ \\
T(\cos(t), \sin(t)) = \begin{pmatrix} 2 \cos(t) + \sin(t) \\ \cos(t) + 2\sin(t) \end{pmatrix} , t \in [0, 2 \pi]
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transformação
A = np.array([[2, 1], [1, 2]])

# Função que aplica a transformação linear
def T(t):
    coss = np.cos(t)
    sen = np.sin(t)
    return A @ [coss, sen] # Multiplicação de matrizes

# Pegando os valores para t
t = np.linspace(0, 2 * np.pi, 100) # t e [0, 2pi]

# Criando a figura
fig = plt.figure()

x, y = T(t)
plt.plot(x, y, label='T')

plt.xlabel('t')
plt.ylabel('T')
```
